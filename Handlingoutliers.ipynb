```python
#boxplot based on 5 number summary: minimum,Q1,median,Q3 and maximum
#Q1 = 25th percentile and so on
#interquertile range (iqr) = Q3-Q1

#generally outliers are values that are
#below Q1 - (1.5*iqr)
#above Q3 + (1.5*iqr)
```

```python
Q1 = wdi_imp['atms_per_100000'].quantile(0.25)
Q3 = wdi_imp['atms_per_100000'].quantile(0.75)
```

```python
print(Q1)
print(Q3)
```

```python
wdi_imp['atms_per_100000'].describe()
```

```python
iqr = Q3-Q1
```

```python
outlier_high = Q3 + (1.5*iqr)
outlier_low = Q1 - (1.5*iqr)
```

```python
print(outlier_high)
print(outlier_low)
```

```python
msk = (wdi_imp['atms_per_100000']<outlier_low) | (wdi_imp['atms_per_100000']>outlier_high)
```

```python
wdi_imp[msk]
```

```python
wdi_imp['atms_per_100000'].hist(bins = 50)
```

```python
p99 = wdi_imp['atms_per_100000'].quantile(0.99)
```

```python
p99
```

```python
wdi_imp.loc[wdi_imp['atms_per_100000']>p99, 'atms_per_100000']
```

```python
wdi_imp.loc[wdi_imp['atms_per_100000']>p99, 'atms_per_100000'] = p99 #capping with a large value to get less skewed result
```
