```python
# Python data visualization libraries:-

# 1. Matplotlib: comprehensive visualization with python
# 2. Pandas: uses matplotlib, good for quick explorations but limited
# 3. Seaborn (statistical data visualization): builds on top of matplotlib and integrates closely with pandas data structure, easy to use and style
# 4. Plotly(Dash): popular for interactive visualization/dashboards.
```

```python
# Distribution of one feature: Seaborn
# 1. Numeric data:
#     histogram: displot()
#     kde: displot()
#     boxplt: catplot()
# 2. Categorical Data:
#     bar(count) plot : catplot()
```

```python
import seaborn as sns
```

```python
sns.displot(data = air, x= 'TEMP') # automatically chooses the bin size based on underlying data
```

```python
sns.displot(data = air, x= 'TEMP', binwidth = 5)
```

```python
sns.displot(data = air, x= 'TEMP', bins = 5, height = 3, aspect = 1) # aspect*height = width
```

```python
sns.displot(data = air, x = 'TEMP', kind = 'kde', height = 3) # default kind is hist
```

```python
sns.displot(data = air, x = 'TEMP', kde = True,rug = True, height = 3, aspect = 2) #rug plot shows each observation with tick mark on x-axis. used to visualize the distribtion
#of data. Useful for smaller datasets
```

```python
sns.catplot(data = air, y = 'TEMP', kind = 'box')
```

```python
sns.catplot(data  = air, x  ='PM2.5_category', kind  ='count',aspect = 2.5)
```

```python
sns.catplot(data  = air, x  ='wd', kind  ='count',aspect = 2.5)
```
