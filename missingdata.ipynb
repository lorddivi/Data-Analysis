```python
wdi = pd.read_pickle('wdi.pkl')
numeric_cols = wdi.select_dtypes(include = 'number').columns
category_cols = wdi.select_dtypes(exclude = 'number').columns
```

```python
wdi[numeric_cols].mean()
```

```python
wdi[numeric_cols].median()
```

```python
wdi[numeric_cols]=wdi[numeric_cols].fillna(wdi[numeric_cols].mean())
```

```python
wdi['alcohol_consumption_per_capita'].value_counts(dropna = False)
```

```python
wdi[category_cols].describe()
```

```python
most_freq = wdi[category_cols].describe().loc['top']
```

```python
wdi[category_cols] = wdi[category_cols].fillna(most_freq)
```

```python
#imputing missing data with models
```

```python
wdi = pd.read_pickle('wdi.pkl')
numeric_cols = wdi.select_dtypes(include = 'number').columns
```

```python
wdi.head(2)
```

```python
from sklearn.experimental import enable_iterative_imputer
```

```python
from sklearn.impute import IterativeImputer
```

```python
iter_imp = IterativeImputer(min_value = wdi[numeric_cols].min(), max_value = wdi[numeric_cols].max())
```

```python
imputed_num_cols = iter_imp.fit_transform(wdi[numeric_cols])
```

```python
wdi[numeric_cols] = imputed_num_cols
```
