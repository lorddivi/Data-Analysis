```python
hero_info
```

```python
hero_info.loc['Superman']
```

```python
hero_info.loc[['Spider-Man','Superman','Wonder Woman']]
```

```python
hero_info.iloc[1:4] #similar to python index slicing this is exclusive
```

```python
hero_info['alignment'] == 'bad'
```

```python
hero_info.loc[hero_info['alignment'] == 'bad'] #boolean condition based selection
```

```python
mask = hero_info['alignment'] == 'bad'
mask
```

```python
hero_info.iloc[mask.values] #iloc doesn't accept condition as argument
```

```python
hero_info[hero_info['alignment'] == 'bad']
```

```python
hero_info[(hero_info['alignment'] == 'bad') & (hero_info['height']>=200)]
```

```python
hero_info[(hero_info['hair color'] == 'No Hair') | (hero_info['hair color'] == 'Blond')]
```

```python
hero_info[hero_info['hair color'].isin(['No Hair','Blond'])]
```

```python
hero_info[~hero_info['alignment'].isin(['good','bad'])] #not operator
```
