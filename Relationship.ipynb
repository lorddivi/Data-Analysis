```python
# #Relationship of two features: seaborn
# 1. Two numeric feature:
#     scatterplot: relplot()/pairplot()/joitplot()
#     correlation heatmap: heatmap()
# 2. One cat and one numeric feature:
#     histogram: displot()
#     line plot: relplot()
#     bar/box/boxen/strip/swarm/violin plots: catplot()
# 3. Two cat feature:
#     count plot: catplot()
```

```python
import pandas as pd
air = pd.read_pickle('air_quality.pkl')
import seaborn as sns
```

```python
sns.relplot(data = air , x = 'TEMP' , y = 'PRES' , kind = 'scatter') # designed to visualize statistical relationships-- scatter default
```

```python
#sns.pairplot(data = air) # this function visualizes the pairwise relationships for numeric features
```

```python
sns.pairplot(data = air[['PM2.5','TEMP','PRES']])
```

```python
sns.jointplot(data = air, x= 'TEMP', y = 'PRES') #by default represents the temp and pres feature using scatter and individual using histogram
```

```python
air.corr()
```

```python
sns.heatmap(data = air.corr(), cmap = 'coolwarm')
```

```python
#relationship of one catg and one numeric feature
```

```python
sns.displot(data = air, x = 'TEMP', hue = 'quarter')
```

```python
sns.displot(data = air, x = 'TEMP', col = 'quarter')
```

```python
sns.relplot(data = air, x = 'quarter', y = 'TEMP', kind = 'line') #plotting the mean of temp for each quarter
```

```python
air.groupby('quarter')['TEMP'].mean()
```

```python
sns.catplot(data = air , x= 'quarter', y = 'TEMP', kind = 'bar', aspect = 2)
```

```python
sns.catplot(data = air , x= 'quarter', y = 'TEMP', kind = 'box', aspect = 2)
```

```python
 sns.catplot(data = air , x= 'quarter', y = 'TEMP', kind = 'boxen', aspect = 2)
```

```python
 sns.catplot(data = air , x= 'quarter', y = 'TEMP', kind = 'strip', aspect = 2)
```

```python
 sns.catplot(data = air[air['PM2.5']>300] , x= 'quarter', y = 'TEMP', kind = 'swarm', aspect = 2)
```
