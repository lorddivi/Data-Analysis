```python
# pandas.pivot_table:
#     create an excel style pivot table as a dataframe
#     agggregate diff statistics based on diff grps(index/row + column)
```

```python
import pandas as pd
student = pd.read_csv('student.csv')
```

```python
student.info()
```

```python
student
```

```python
pd.pivot_table(student,index = 'sex') # mean
```

```python
pd.pivot_table(student,index = ['sex','internet'])
```

```python
pd.pivot_table(student,index = ['sex','internet'],values = 'score')
```

```python
pd.pivot_table(student,index = ['sex','internet'],values = 'score',aggfunc = ['min','median','max'])
```

```python
pd.pivot_table(student,index = ['sex','internet'],values = 'score',columns = 'studytime')
```

```python
pd.pivot_table(student,index = ['sex','internet'],values = 'score',columns = 'studytime',fill_value = -999)
```

```python
pd.pivot_table(student,index = ['sex','internet'],values = ['score','age'],columns = 'studytime',fill_value = -999)
```

```python
pd.pivot_table(student,index = 'sex',values = ['score','age'],columns = ['studytime','internet'],fill_value = -999)
```

```python
pd.pivot_table(student, index = 'familysize', columns  = 'sex',values = 'score')
```

```python
pd.pivot_table(student, index = 'familysize', columns  = 'sex',values = 'score',margins = True,margins_name = 'Avg score')
```

```python
student[student['sex'] == 'F'].mean()
```

```python
# aggfunc works same as agg
```

```python
pd.pivot_table(student, 
               index = 'studytime',
               columns = 'sex',
               values = ['age','score'], 
               aggfunc = {'age':['max','min'],'score':'median'})
```

```python
def rangef(s):
    return s.max()-s.min()
```

```python
pd.pivot_table(student, 
               index = 'studytime',
               columns = 'sex',
               values = 'score',
               aggfunc = rangef)
```

```python
pd.pivot_table(student, 
               index = 'studytime',
               columns = 'sex',
               values = 'score',
               aggfunc = lambda s:s.max()-s.min())
```

```python
# summary:
# index: keys to grp by on the pivot table index/row
# columns: keys to grp by o the pivot table column
# values: column to aggregate
# aggfunc: functions to use for aggregation
# margins = True
# fill_value
```
